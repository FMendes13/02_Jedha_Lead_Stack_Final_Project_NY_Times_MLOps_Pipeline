# Library

```python
import pandas as pd 
import requests
from dotenv import load_dotenv
import os
from datetime import datetime
```

# News API

```python
#News API - https://newsapi.org/
load_dotenv()
API_KEY= os.getenv('API_KEY')

url= f'https://newsapi.org/v2/top-headlines?country=us&apiKey={API_KEY}'
response=requests.get(url)
response
```

```python
response=response.json()
display (response)
```

```python
df_newsapi=pd.json_normalize(response['articles'])
```

```python
df_newsapi.to_json('df_newsapi.json')
```

```python
df_newsapi=pd.read_json('df_newsapi.json')
df_newsapi
```

```python
df_newsapi.shape
```

# NY Times

```python
#NY Times
load_dotenv()
NY_Times_Key= os.getenv('NY_Times_Key')

url= f'https://api.nytimes.com/svc/news/v3/content/section-list.json?api-key={NY_Times_Key}'
response=requests.get(url)
response
```

```python
response.json()
```

```python
url= f'https://api.nytimes.com/svc/news/v3/content/all/business.json?api-key={NY_Times_Key}'
response=requests.get(url)
response
```

```python
data=response.json()
data
```

```python
results = data.get('results', [])
df_ny_times=pd.json_normalize(results)

df_ny_times.to_json('df_ny_times.json')
df_ny_times=pd.read_json('df_ny_times.json')

df_ny_times
```

```python
df_ny_times.shape
```

```python
current_date = datetime.now().strftime('%Y-%m-%d_%H-%M')
filename = f'df_ny_times_{current_date}.csv'
df_ny_times.to_csv(filename, index=False)
```
